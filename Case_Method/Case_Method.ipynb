```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import math
from PIL import Image as im
```

```python
image = cv.imread(dir + 'Case 2.jpg')
```

```python
# Improve visibility by adjusting contrast and brightness
enhanced_image = cv2.convertScaleAbs(image, alpha=3.5, beta=50)

# Convert BGR to RGB for display
enhanced_image_rgb = cv2.cvtColor(enhanced_image, cv2.COLOR_BGR2RGB)

# Convert the original image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display the original, grayscale, and enhanced images
plt.figure(figsize=(15, 6))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

# Grayscale Image
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

# Enhanced Image
plt.subplot(1, 3, 3)
plt.imshow(enhanced_image_rgb)
plt.title("Enhanced Image")
plt.axis('off')

plt.show()
```
