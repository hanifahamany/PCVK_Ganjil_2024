```python
def js_to_image(js_reply):
  image_bytes = b64decode(js_reply.split(',')[1])
  jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
  img = cv2.imdecode(jpg_as_np, flags=1)

  return img


def bbox_to_bytes(bbox_array):
  bbox_PIL = PIL.Image.fromarray(bbox_array, 'RGBA')
  iobuf = io.BytesIO()
  bbox_PIL.save(iobuf, format='png')
  bbox_bytes = 'data:image/png;base64,{}'.format((str(b64decode(iobuf.getvalue()), 'utf-8')))

  return bbox_to_bytes
```

```python
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality){
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)

  data = eval_js('takePhoto({})'.format(quality))
  img = js_to_image(data)
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  print(gray.shape)
  faces = face_cascade.detectMultiScale(gray)
  for(x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255, 0, 0),2)

  cv2.imwrite(filename, img)

  return filename
```

7. Lakukan segmentasi karakter pada KTP.

```python
from google.colab.patches import cv2_imshow
# Baca gambar KTP
image = cv2.imread('/content/drive/MyDrive/PCVK_2024/Week_11_Hanifah/Object Detection/ktp3.png')

# Praproses gambar
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blur, 50, 150)

# Binarisasi
ret, binary = cv2.threshold(edges, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Segmentasi Karakter
contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Loop melalui kontur dan gambar kotak di sekitar karakter
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Tampilkan hasil
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

8. Lakukan cropping pada bagian NIK.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
# Baca gambar KTP
image = cv2.imread('/content/drive/MyDrive/PCVK_2024/Week_11_Hanifah/Object Detection/ktp3.png')

# Praproses gambar
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (5, 5), 0)
edges = cv2.Canny(blur, 50, 150)

# Binarisasi
ret, binary = cv2.threshold(edges, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Segmentasi Karakter
contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Inisialisasi variabel untuk menyimpan area maksimum NIK
max_nik_area = 0
max_nik_contour = None

# Loop melalui kontur dan cari kotak yang sesuai dengan kriteria NIK
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)

    # Tentukan kriteria untuk NIK (misalnya, dengan asumsi NIK berada di bagian atas gambar)
    if y < 100 and h > 15:
        # Hitung luas kontur
        area = cv2.contourArea(contour)

        # Perbarui jika luas kontur lebih besar dari yang sebelumnya
        if area > max_nik_area:
            max_nik_area = area
            max_nik_contour = contour

# Pastikan kontur NIK ditemukan sebelum melakukan cropping
if max_nik_contour is not None:
    # Buat bounding box untuk kontur NIK
    x, y, w, h = cv2.boundingRect(max_nik_contour)

    # Cropping hanya bagian NIK
    nik_cropped = image[y:y + h, x:x + w]

    # Simpan hasil cropping
    cv2.imwrite('path/to/save/nik_cropped.jpg', nik_cropped)

    # Tampilkan hasil cropping
    cv2_imshow(nik_cropped)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Kontur NIK tidak ditemukan.")
```
